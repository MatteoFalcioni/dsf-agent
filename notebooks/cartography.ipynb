# Downloading Bologna's Catography

Here we just download the cartography, edges and nodes file.

> NOTE: download libtbb-dev with `sudo apt install libtbb-dev`.

```python
from dsf import get_cartography

gdf_edges, gdf_nodes = get_cartography(place_name="Bologna", return_type="gdfs")

gdf_edges.to_csv("../data/bologna_edges.csv")
gdf_nodes.to_csv("../data/bologna_nodes.csv")
```
